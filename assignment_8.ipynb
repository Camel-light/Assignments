9. What is an embedding matrix?
*An embedding matrix is a matrix of weights that maps each word in a vocabulary to a vector of real numbers. The embedding matrix is used in natural language processing (NLP) tasks such as text classification and sentiment analysis. The matrix is typically initialized randomly and then learned during training using backpropagation. The size of the embedding matrix is determined by the number of words in the vocabulary and the number of dimensions in the vector space.*

10. What is the relationship between an embedding and a matrix of one-hot-encoded vectors?
*An embedding is a dense vector representation of a word that captures its meaning. In contrast, one-hot encoding is a sparse vector representation of a word that represents its presence or absence in a vocabulary. An embedding matrix maps each word in the vocabulary to an embedding vector, while a matrix of one-hot-encoded vectors maps each word to a sparse vector with only one non-zero element. The relationship between the two is that the embedding matrix can be used to transform one-hot-encoded vectors into dense embeddings.*

11. Why do we need `Embedding` if we could use one-hot-encoded vectors for the same thing?

*One-hot-encoded vectors are sparse and do not provide insight into the relationships between words. Word embeddings, on the other hand, provide a continuous and dense representation that can be used to measure similarity and distance between words. Embeddings can also capture the context of a word in a sentence or document. In addition, embeddings can be learned during training using backpropagation. This allows the model to learn the optimal representation for each word in the vocabulary.*

26. What is the "bootstrapping problem" in collaborative filtering?

*The bootstrapping problem in collaborative filtering refers to the problem of how to make recommendations for new users or items that have no ratings yet.*

27. How could you deal with the bootstrapping problem for new users? For new movies?

*One way to deal with the bootstrapping problem for new users is to use content-based filtering. This method uses the attributes of the items that the user has rated to recommend new items.*

*Another way is to use collaborative filtering. Collaborative filtering uses similarities between users and items simultaneously to provide recommendations. This allows for serendipitous recommendations; that is, collaborative filtering models can recommend an item to user A even though no other users have rated both that item and user A’s other items.*

````python
trainset, testset = train_test_split(data, test_size=0.2)

algo = SVD(n_factors=100, lr_all=0.008, reg_all=0.02, n_epochs=25, biased=True)
algo.fit(trainset)

predictions = algo.test(testset)

user_ids = np.unique([uid for (uid, _, _) in testset])[:10]

plt.figure(figsize=(12, 9))

actual_ratings_agg = []
predicted_ratings_agg = []

for uid in user_ids:
    user_testset = [rating for rating in testset if rating[0] == uid]
    user_predictions = [pred.est for pred in predictions if pred.uid == uid]
    
    actual_ratings_agg.append(np.mean([r for (_, _, r) in user_testset]))
    predicted_ratings_agg.append(np.mean(user_predictions))

bar_width = 0.39

x = np.arange(len(user_ids))

plt.figure(figsize=(12, 9))
plt.bar(x, actual_ratings_agg, width=bar_width, label='Actual R', alpha=0.8)
plt.bar(x + bar_width, predicted_ratings_agg, width=bar_width, label='Predicted Rat.', alpha=0.8)

plt.xticks(x + bar_width / 2, user_ids)

plt.xlabel('User IDs')
plt.ylabel('Ratings')
plt.title('Average Actual vs Predicted Ratings for Selected Users')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.9)
plt.tight_layout()
plt.show()

rmse = accuracy.rmse(predictions)
````
